```python
# Create the sales_state DataFrame
sales_state = df.groupby('State')['Orders'].sum().reset_index()

plt.figure(figsize=(10, 12))
ax = sns.barplot(y='State', x='Orders', hue='State', data=sales_state, palette='coolwarm', dodge=False, legend=False)

# Add value labels
for bar in ax.patches:
    ax.text(bar.get_width() + 5, bar.get_y() + bar.get_height()/2,
            int(bar.get_width()), va='center')

plt.title('Total Orders by State')
plt.tight_layout()
plt.show()
```

```python
ax = sns.countplot(x="Marital_Status",hue='Marital_Status',data=df, palette="deep")
plt.show()
```

```python
# Clean and prepare data
df['Marital_Status'] = df['Marital_Status'].astype(str).str.strip()
df = df.dropna(subset=['Marital_Status'])

# Group and sort
sales_state = df.groupby('Marital_Status', as_index=False)['Amount'].sum()
sales_state = sales_state.sort_values(by='Amount', ascending=False)

# Plot
plt.figure(figsize=(8, 5))
ax = sns.barplot(
    y='Marital_Status',
    x='Amount',
    hue='Marital_Status',          # fixes the warning
    data=sales_state,
    palette='coolwarm',
    dodge=False,
    legend=False                   # hides redundant legend
)

# Add value labels
for bar in ax.patches:
    ax.text(bar.get_width() + 10000,
            bar.get_y() + bar.get_height() / 2,
            int(bar.get_width()),
            va='center')

plt.title('Total Spent according to Marital Status')
plt.xlabel('Total Amount')
plt.ylabel('Marital Status')
plt.tight_layout()
plt.show()
```

```python
from the above graph we can say that most of the buyers are married 
```

```python
plt.figure(figsize=(8, 10))  # Taller plot for horizontal bars

# Use `hue="Occupation"` and disable the legend
ax = sns.countplot(y="Occupation", hue="Occupation", data=df, palette="deep", legend=False)

plt.title("Count of Customers by Occupation")
plt.tight_layout()
plt.show()
```

```python
# Grouping by Occupation and summing the Amount
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(8, 6))  # Wider for clearer labels

# Assign hue and disable legend to avoid warning
ax = sns.barplot(y='Occupation', x='Amount', hue='Occupation', data=sales_occupation,
                 palette='coolwarm', dodge=False, legend=False)

plt.title('Total Spent According to Occupation')
plt.xlabel('Total Amount')
plt.ylabel('Occupation')
plt.tight_layout()
plt.show()
```

from above graph we can say that most of the buyers is from IT sector , healthcare, aviation

```python
plt.figure(figsize=(12, 6))  # Make the figure wider

# Add hue=x and disable legend to avoid FutureWarning
ax = sns.countplot(x="Product_Category", hue="Product_Category", data=df, palette="deep", legend=False)

plt.xticks(rotation=45, ha='right')  # Rotate labels for better fit
plt.title("Count of Product Categories")
plt.tight_layout()
plt.show()
```

```python
# Grouping and sorting
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum()
sales_state = sales_state.sort_values(by='Amount', ascending=False)

# Plotting
plt.figure(figsize=(5, 6))  # Slightly wider for clarity
ax = sns.barplot(
    y='Product_Category',
    x='Amount',
    hue='Product_Category',        #  to avoid FutureWarning
    data=sales_state,
    palette='coolwarm',
    dodge=False,
    legend=False                   # hides duplicate legend
)

plt.title('Total Amount Spent by Product Category')
plt.xlabel('Total Amount')
plt.ylabel('Product Category')
plt.tight_layout()
plt.show()
```

most sold products are food , clothing

conclusion: married women age groups 26-35 years from up, maharastra and karnataka working in IT, hralthcare, aviation are more likely to buy products from food, clothing and electronic category
